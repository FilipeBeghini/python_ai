### Case: Credit Score of Clients

```python
# 1. Get database
import pandas as pd

db = pd.read_csv ("clientes.csv") # create database as db
display(db)
```

```python
# 2. Papare the database for the AI 
display(db.info())
# the database sample is already padronized so no need to ajust for errors on it
    # db=db.dropna() # NaN - Not a Number - Empty info (dropna = drop NaN = drop empty info)
    # display(db.info())
    # db = db.drop(columns="ID_cliente")
    # display(db)

# AI only reads Dtype numbers only, so we need to conver Objects to numbers for all necessary data that the AI needs to lear.

# int -> integral numbers
# float -> decimal numbers
# object -> texts

# credit score:

# Good = Good
# Standard = OK
# Poor = Bad

```

```python
# Lable Encoder
from sklearn.preprocessing import LabelEncoder

# Profissao
    # mecanico - 1
    # professor - 2
    # artista - 3
    # advogado - 4
    # medico - 5
    # bombeiro - 6
    # etc.

en_profissao = LabelEncoder()
db["profissao"] = en_profissao.fit_transform(db["profissao"])

# mix_credito
en_mix_cred = LabelEncoder()
db["mix_credito"] = en_mix_cred.fit_transform(db["mix_credito"])

# comportamento_pagamento
en_comp_pag = LabelEncoder()
db["comportamento_pagamento"] = en_comp_pag.fit_transform(db["comportamento_pagamento"])

# x = the column that the AI will use to learn 
# y = the column that the AI will use to be the prediction (no need to be numbers only)

y = db["score_credito"] # use the specified column to lear what is the prediction
x = db.drop(columns=["score_credito", "id_cliente"]) # use all columns but not the one we need to predict nor irrelevant columns

# devide the database for train data and test data
# so we will have an x_train, x_test, y_train, y_test -> means that we need to alocate some percentage of the data we separated
#  for the AI lear and it is going to use for test the quality of the predictions

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3) # arrange the data for train and for test by 70% - 30%
                                                                         # if doesn't define test_size it will devide automatically.
```

```python
# 3. Create an AI model -> Grade system/Credit score (Good, OK, Bad)
# Decision tree -> Randomforest
# Nearest Neighbors -> KNN

# importing the AI
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# creating AI
decision_tree_model = RandomForestClassifier()
knn_model = KNeighborsClassifier()

# train AI
decision_tree_model.fit(x_train, y_train)
knn_model.fit(x_train, y_train)
```

```python
# 4. Avail which is the best AI model to use
decision_tree_prediction = decision_tree_model.predict(x_test)

knn_predict = knn_model.predict(x_test)

# Calculating accuracy
from sklearn.metrics import accuracy_score

print("Decision tree model accuracy:")
display(accuracy_score(y_test, decision_tree_prediction))
print("KNN model accuracy:")
display(accuracy_score(y_test, knn_predict))

# we can see that the decidion tree AI has a better accuracy, so we gonna use it.
```

```python
# 5. Make new credit score predictions
# get new database and use the chosen model of AI to predict the credit score
new_clients_db = pd.read_csv("novos_clientes.csv")
display(new_clients_db)

# preparing new data for the AI using the previus encoders:
# profissao encoding object to number
new_clients_db["profissao"] = en_profissao.fit_transform(new_clients_db["profissao"])
# mix_credito encoding object to number
new_clients_db["mix_credito"] = en_mix_cred.fit_transform(new_clients_db["mix_credito"])
# comportamento_pagamento encoding object to number
new_clients_db["comportamento_pagamento"] = en_comp_pag.fit_transform(new_clients_db["comportamento_pagamento"])

# predicting using the new data using the decision tree AI model
decision_tree_prediction_new_data = decision_tree_model.predict(new_clients_db)
display(decision_tree_prediction_new_data)

# joblib / picket or something like that to save the variable (model AI) into a file so you can use it anywhere 
```
